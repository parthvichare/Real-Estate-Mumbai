```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
webpage=requests.get('https://www.magicbricks.com/flats-in-mumbai-for-sale-pppfs').text
```

```python
soup= BeautifulSoup(webpage,'lxml')
soup
```

```python
#To find out the name of the properties
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
for i in soup.find_all('p'):
    print(f'{i.text.strip()}\n')
```

```python
#Carpet Area
for i in soup.find_all(class_='mb-srp__card__summary--value'):
    print(f'{i}\n')

#     soup.find_all('div',class_='mb-srp__card__summary--value')
```

```python
a=soup.find_all('div', class_='mb-srp__card__summary--value')
len(a)
```

```python
a = soup.find_all('div', class_='mb-srp__card__summary--value', text=lambda text: text and 'sqft' in text)
for i in a:
    print(i.text.strip())
```
