```python
address_list=[]
for url in feature['Link']:
    addres = []
    try:
        webpage = requests.get(url).text
        soup = BeautifulSoup(webpage, 'html.parser')
        for i in soup.find_all(class_='proj-info__value'):
            addres.append(i.text.strip())
        address_list.append(addres)
        
    except Exception as e:
        print(f"Error fetching or parsing data from {url}: {str(e)}")

# Now url_amenities_list contains a list of amenities for each non-null URL
df = pd.DataFrame({'Address': address_list})
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

address_list = []
pincode_list = []
towers=[]

for url in feature['Link']:
    try:
        webpage = requests.get(url).text
        soup = BeautifulSoup(webpage, 'html.parser')
        property_details = soup.find_all('div', class_='proj-info__data__block')

        address = None
        pincode = None

#Get the Idea of Html Structure
        for detail in property_details:
            title = detail.find('div', class_='proj-info__data__title')
            value = detail.find('div', class_='proj-info__data__value')
#             value_1=detail.find('div', class_='proj-info__data__value')

            if title and value:
                if title.text == 'Address':
                    address = value.text.strip()
                elif title.text == 'Pincode':
                    pincode = value.text.strip()

        # Append the data to the appropriate lists
        address_list.append(address)
        pincode_list.append(pincode)

    except Exception as e:
        print(f"Error processing {url}: {e}")

# Create a DataFrame if needed
df = pd.DataFrame({'Address': address_list, 'Pincode': pincode_list})
print(df)
```

```python
address_list = []
pincode_list = []  # Corrected variable name for pincode list

for url in feature['Link']:
    try:
        webpage = requests.get(url).text
        soup = BeautifulSoup(webpage, 'html.parser')
        property_details = soup.find_all('div', class_='proj-info__data')
        property_functions = soup.find_all('div', class_='sub-wrap')
        
        for i, info in zip(property_details, property_functions):  # Use a different variable name for the second loop
            address_element = i.find(class_='proj-info__data__value')
            address_list.append(address_element.text.strip() if address_element else np.nan)
            
            pincode_label = info.find('div', class_='proj-info__data__label', text='Pincode')
            pincode_value = pincode_label.find_next('div', class_='proj-info__data__value').text.strip() if pincode_label else np.nan
            pincode_list.append(pincode_value)
            
    except Exception as e:
        print(f"Error on page {url}: {str(e)}")

df = pd.DataFrame({'address': address_list, 'Pincode': pincode_list})
```
