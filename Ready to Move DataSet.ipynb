```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Data to store in List
name = []
society_name = []
price = []
sqft = []
carpet_area = []
furnished_status = []
bathroom = []
bedrooms = []
floor = []
link = []
balcony=[]

for page_number in range(1,160):
    url = f'https://www.magicbricks.com/ready-to-move-flats-in-mumbai-pppfs?page={page_number}'

    try:
        webpage = requests.get(url).text
        soup = BeautifulSoup(webpage, 'html.parser')

        property_details = soup.find_all('div', class_='mb-srp__card')
        property_info = soup.find_all('div', class_='mb-srp__card__summary__list')

        for i, info in zip(property_details, property_info):
            name.append(i.find('h2').text.strip())
            
            # Extracting Property Details Link
            link_element = i.find('a', href=True)
            link.append(link_element['href'].strip() if link_element and 'href' in link_element.attrs else np.nan)

            society_element = i.find(class_='mb-srp__card__society--name')
            society_name.append(society_element.text.strip() if society_element else np.nan)
            
            price_element = i.find(class_='mb-srp__card__price--amount')
            sqft_element = i.find(class_='mb-srp__card__price--size')

            price.append(price_element.text.strip() if price_element else np.nan)
            sqft.append(sqft_element.text.strip() if sqft_element else np.nan)

            # Extracting Carpet Area
            carpet = info.find('div', class_='mb-srp__card__summary--label', text='Carpet Area')
            carpet_area_value = carpet.find_next('div', class_='mb-srp__card__summary--value').text.strip() if carpet else np.nan
            carpet_area.append(carpet_area_value)
            
            furnished_label = info.find('div', class_='mb-srp__card__summary--label', text='Furnishing')
            furnished_value = furnished_label.find_next('div', class_='mb-srp__card__summary--value').text.strip() if furnished_label else np.nan
            furnished_status.append(furnished_value)
            
            floor_label = info.find('div', class_='mb-srp__card__summary--label', text='Floor')
            floor_value = floor_label.find_next('div', class_='mb-srp__card__summary--value').text.strip() if floor_label else np.nan
            floor.append(floor_value)
            
            bathroom_label = info.find('div', class_='mb-srp__card__summary--label', text='Bathroom')
            bathroom_value = bathroom_label.find_next('div', class_='mb-srp__card__summary--value').text.strip() if bathroom_label else np.nan
            bathroom.append(bathroom_value)
            
            balcony_label = info.find('div', class_='mb-srp__card__summary--label', text='Balcony')
            balcony_value = balcony_label.find_next('div', class_='mb-srp__card__summary--value').text.strip() if balcony_label else np.nan
            balcony.append(balcony_value)


    except Exception as e:
        print(f"Error on page {page_number}: {str(e)}")

# Manipulating the string Sqft
df = pd.DataFrame({'Name': name,
                   'Link': link,
                   'Society': society_name, 
                   'Price': price, 'Sqft': sqft, 
                   'Carpet': carpet_area, 
                   'Furnished': furnished_status, 
                   'Floor': floor,
                   'balcony':balcony,
                   'bathroom':bathroom})
df['Sqft'] = df['Sqft'].apply(lambda x: x[0:15] if x is not np.nan else np.nan)
df['Bedroom'] = df['Name'].str.extract(r'(\d)')
df
```
