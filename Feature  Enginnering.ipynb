```python
#Create 2Nan Columns for filling the buildup and SuperBuild Area
# Define the number of rows
num_rows = 3400

# Create a dictionary to hold the columns
columns = {}

# Specify the number of columns you want
num_columns =2 # Change this to the desired number of columns

# Create columns with all NaN values
for i in range(num_columns):
    column_name = f'Column_{i+1}'
    columns[column_name] = [np.nan] * num_rows

# Create a DataFrame using the dictionary of columns
columns = pd.DataFrame(columns)
```

```python
columns=columns.rename(columns={'Column_1':'Buildup','Column_2':'SuperBuild_area'})
```
