```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('Feature_flats.xlsx')
```

```python
#Filing Missing Value of the Buildup Area
```

```python
one=df[df['bedroom']==1]['Buildup']
two=df[df['bedroom']==2]['Buildup']
three=df[df['bedroom']==3]['Buildup']

#Calculating avg mean
onebhk_mean=one.mean()
twobhk_mean=two.mean()
threebhk_mean=three.mean()

#Filling Missing rows with mean
one=one.fillna(onebhk_mean)
two=two.fillna(twobhk_mean)
three=three.fillna(twobhk_mean)

# concat all the missing values in the buildup Column
df['Buildup']=pd.concat([one,two,three])
```

```python
#Filling Missing Value of the SuperBuild Area
```

```python
one=df[df['bedroom']==1]['SuperBuild_area']
two=df[df['bedroom']==2]['SuperBuild_area']
three=df[df['bedroom']==3]['SuperBuild_area']

#Calculating avg mean
onebhk_mean=one.mean()
twobhk_mean=two.mean()
threebhk_mean=three.mean()

#Filling Missing rows with mean
one=one.fillna(onebhk_mean)
two=two.fillna(twobhk_mean)
three=three.fillna(twobhk_mean)

# concat all the missing values in the buildup Column
df['SuperBuild_area']=pd.concat([one,two,three])
```
